```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ex=sns.load_dataset("exercise")
ex.head()
```

```python
fig=plt.figure(figsize=(12,8))
plt.suptitle("Exercise Dataset  Pulse Analysis Dashboard", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplot(2,3,1)
sns.barplot(x='kind',y='pulse',data=ex)
plt.title('Average pulse by exercize Type')
plt.subplot(2,3,2)
plt.title('Normal pulse ki range')
sns.histplot(x='pulse',kde=True,data=ex)
plt.subplot(2,3,3)
plt.title('time ke sath pulse ka effect')
sns.lineplot(x='time',y='pulse',hue='kind',data=ex)
plt.subplot(2,3,4)
plt.title('Low-fat diet aur no-fat diet me difference')
sns.boxplot(x="diet", y="pulse", data=ex)
plt.subplot(2,3,5)
plt.tight_layout(rect=[0, 0, 1, 0.95])
sns.barplot(x="kind", y="pulse", hue="diet", data=ex)
plt.title('Diet + kind milkar pulse ko kaise affect karta hai')
plt.subplot(2,3,6)
sns.scatterplot(x="time", y="pulse", hue="kind", data=ex)


```
