# Violin plot

```python
import plotly.express as px
```

```python
df=px.data.tips()
df.head()
```

```python
px.violin(df,y="total_bill")
```

```python
px.violin(df,y="total_bill",box=True,points="all")
```

```python
px.violin(df,y="total_bill",box=True,points="outliers")
```

```python
fig=px.violin(df,y="tip",color="time",violinmode="overlay",hover_data=df.columns)
fig
```

```python
px.strip(df,x="day",y="tip")
```

```python
fig=px.violin(df,y="total_bill")
fig.update_traces(quartilemethod="exclusive")
```
