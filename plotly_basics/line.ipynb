# Line plot

```python
import plotly.express as px
import pandas as pd
```

```python
df=px.data.gapminder()
df.head()
```

```python
df=px.data.gapminder().query("country=='Canada'")
df.head()
```

```python
fig=px.line(df,x="year",y="lifeExp",color="country",title="life exp in canada")
fig
```

```python
df=px.data.gapminder().query("continent=='Oceania'")
df.head()
```

```python
df.tail()
```

```python
fig=px.line(df,x="year",y="lifeExp",color="country",title="life exp in canada",markers=True)
fig
```
