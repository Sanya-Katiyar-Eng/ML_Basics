# Bar plot

```python
import plotly.io as pio
pio.renderers.default = "browser"

```

```python
import plotly.express as px
import pandas as pd
df=pd.read_csv(r"C:\Users\user\Downloads\Iris.csv")
print(df.head())

px.bar(df,x='Species',color="Species")
```

if we want to add difrent color

```python
px.bar(df,x='Species',color="Species",color_discrete_sequence=["red","green","blue"])
```

size of species

```python
data=df.groupby(["Species"]).size().rename("count").reset_index()
data.head()
```

```python
px.bar(data,x='Species',y='count',color="Species",text="count")
```
