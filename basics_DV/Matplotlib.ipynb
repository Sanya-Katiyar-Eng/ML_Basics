*Object oriented way*


**matplotlib allows the aspect ratio,DPI and figure size to be specified when the figure object is created . you can use the ***"figsize"*** and dpi ***"keyword"*** arguments.**
**figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch(pixel per inch)**

#### 

```python
fig=plt.figure(figsize=(12,8),dpi=100)
#fig.add_axes(left,bottom,width,height)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
plt.plot(x,y)
plt.title("axes1")
axes2=fig.add_axes([0.2,0.5,0.3,0.3])
plt.plot(x,x)
plt.title("axes2")
```

*OPTIONAL DESIGNING*

```python
import numpy as np
xaxis=np.linspace(0,5,5)
yaxis = np.ones_like(xaxis)
fig, ax=plt.subplots(figsize=(12,6))
ax.plot(xaxis,xaxis+1,color="red",linewidth=1)
ax.plot(xaxis,xaxis+2,color="red",linewidth=2)
ax.plot(xaxis,xaxis+3,color="red",linewidth=3)
ax.plot(xaxis,xaxis+4,color="red",linewidth=4)
#possible linestyle options '-','--',':'-.'steps'
ax.plot(xaxis,xaxis+5,color="green",lw=3,linestyle='-')
ax.plot(xaxis,xaxis+6,color="green",lw=3,ls=':')
ax.plot(xaxis,xaxis+7,color="green",lw=3,ls='-.')
# possible marker symbols '.': 'point',       # small dot',': 'pixel',       # tiny pixel 'o': 'circle',      # circle    'v': 'triangle_down',    '^': 'triangle_up',   '<': 'triangle_left',    '>': 'triangle_right',  '1': 'tri_down',    # triplet down  '2': 'tri_up',  '3': 'tri_left',  '4': 'tri_right', 's': 'square', 'p': 'pentagon', '*': 'star'  'h': 'hexagon1',  'H': 'hexagon2',   '+': 'plus',  'x': 'x'  'D': 'diamond',  'd': 'thin_diamond',   '|': 'vline',       # vertical lin   '_': 'hline'        # horizontal line
ax.plot(xaxis,xaxis+8,color="blue",lw=3,ls='-',marker='+')
ax.plot(xaxis,xaxis+9,color="blue",lw=3,ls='-',marker='o')
ax.plot(xaxis,xaxis+10,color="blue",lw=3,ls='-',marker='s')
ax.plot(xaxis,xaxis+11,color="blue",lw=3,ls='-',marker='1')
ax.plot(xaxis,xaxis+12,color="blue",lw=3,ls='-',marker=2)
ax.plot(xaxis,xaxis+13,color="blue",lw=3,ls='-',marker='>')
ax.plot(xaxis,xaxis+14,color="blue",lw=3,ls='-',marker='*')
ax.plot(xaxis,xaxis+15,color="blue",lw=3,ls='-',marker='>')
```
