we use distribution plots to visualize the distribution of quanntitative data .

let's discuss some plots that allow us to visualize the distribution of a data set.these plots are :
* distplot / histplot
* jointplot
* pairplot
* rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=sns.load_dataset("tips")
data
```

distplot/histplot

```python
plt.subplot(1,2,1)
sns.histplot(data['total_bill'],bins=20,kde='True')
plt.title("total bill")
#kde=how skew data is
#bins=
plt.subplot(1,2,2)
sns.histplot(data['tip'],bins=20,kde='True')
plt.title("tip")
plt.tight_layout()
plt.show()
```

jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='hist')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='resid')
```

```python
sns.pairplot(data,hue='sex',palette='rainbow')
```

```python
sns.rugplot(data['total_bill'])
```
