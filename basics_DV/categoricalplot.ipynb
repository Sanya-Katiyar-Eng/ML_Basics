## Categorical Data plots

now let's discuss using seaborn to plot categorical data ! these are a few main plot types for this .

boxplot
violinplot
stripplot
swarmplot
barplot
countplot
let's go through example

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data
```

```python
sns.countplot(x=data['sex'],hue=data['smoker'])
```

```python
sns.boxplot(x=data['sex'])
```

```python
sns.stripplot(x=data['sex'],y=data['time'])
```

```python
sns.swarmplot(x=data['sex'],y=data['time'])
```

```python
sns.barplot(x=data['sex'],y=data['time'])
```
