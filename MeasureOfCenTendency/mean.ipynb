# Mean

* 1.using formula

```python
l=[10,20,30,40,50,60]
m=sum(l)/len(l)
print("mean :",m)
```

*2.using numpy*

```python
import numpy as np
l=[10,20,30,40,50,60,]
mean=np.mean(l)
print("Mean : ",mean)
```

*3. using pandas*

```python
import pandas as pd 
l=[10,20,30,40,50,60]
m=pd.Series(l)
print("Mean : ",m.mean())
```

*4. using statistic*

```python
import statistics as stats 
l=[10,20,30,40,50,60]
mean=stats.mean(l)
print("Mean :",mean)
```

*5. Mean through DATA SET iris*

```python
import seaborn as sns
import numpy as np
d=sns.load_dataset("iris")
print(d)
print("Mean :",np.mean(d['petal_length']))
```
