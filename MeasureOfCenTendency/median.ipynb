# Median

*1. using formula*

```python
l=[10,20,30,40,50,60]
n=len(l)
if(n%2==0):
    num=int(n/2)
    mid=(l[num]+l[num-1])/2
    print("Median :",l[mid])
else:
    num=int((n+1)/2)
    print("Median :",l[num])
```

*2. using numpy*

```python
import numpy
l=[10,20,30,40,50,60]
m=numpy.median(l)
print("Median :",m)
```

*3. using pandas*

```python
import pandas
l=[10,20,30,40,50,60]
s=pandas.Series(l)
m=s.median()
print("Median :",m)
```

*4. using statistics*

```python
import statistics
l=[10,20,30,40,50,60]
m=statistics.median(l)
print("Median :",m)
```

*5. using DATA SET*

```python
import seaborn as sns
import numpy as np
d=sns.load_dataset("iris")
print(d)
print("Median : ",np.median(d['sepal_length']))
```
