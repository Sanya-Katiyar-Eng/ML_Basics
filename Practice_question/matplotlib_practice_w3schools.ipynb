```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,10,5)
fig=plt.figure(figsize=(4,3))
plt.title("Draw a line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,x,scalex=True,scaley=True)
```

```python
x=[1,2,3]
y=[2,4,1]
plt.figure(figsize=(4,3))
plt.title("Sample graph !")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
```
