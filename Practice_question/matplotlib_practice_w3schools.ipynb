```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,10,5)
fig=plt.figure(figsize=(4,3))
plt.title("Draw a line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,x,scalex=True,scaley=True)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
data={'Date':['3 oct','4 oct','5 oct','6 oct','7 oct'],
          'Open':['774.25','776.030029','779.309998','779','779.659973'],
          'High':['776.065002','778.710022','782.070007','780.47998','779.659973'],
          'Low':['769.5','772.890015','775.650024','775.539978','770.75'],
          'Close':['772.559998','776.429993','776.469971','776.859985','775.080017']}
df=pd.DataFrame(data)
df['Open']=df['Open'].astype(float)
df['Close']=df['Close'].astype(float)
df['Low']=df['Low'].astype(float)
df['High']=df['High'].astype(float)
#df['Date']=df['Date'].astype(float)
plt.plot(df['Date'],df['Open'],label='open')
plt.plot(df['Date'],df['Close'],label='close')
plt.plot(df['Date'],df['Low'],label='low')
plt.plot(df['Date'],df['High'],label='high')
plt.legend()
plt.show()
```

```python
x=[10,20,30]
y=[20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.figure(figsize=(4,3))
plt.plot(x,y,lw=2,ls='-.',label='line_width_3')
plt.plot(x2,y2,lw=2,ls=':',label='line_width_5')
plt.legend()
```

```python
plt.plot(x,y,linewidth=3,label='line1_dotted')
plt.plot(x2,y2,linewidth=5,label='line2_dotted')
plt.legend()
```
