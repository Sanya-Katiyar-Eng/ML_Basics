```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data with an outlier
marks = [45, 48, 50, 52, 49, 47, 51, 46, 50, 100]

# Create box plot
plt.figure(figsize=(6,4))
sns.boxplot(y=marks)

# Add title and label
plt.title("Box Plot of Student Marks (with Outlier)")
plt.ylabel("Marks")

plt.show()
```

It meanse the outlier shown with dot .

# Removing Outlier

```python
import pandas as pd
import numpy as np
arr=np.array(marks)
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outlier=arr[(arr>upper_bound) | (arr<lower_bound)]
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outlier)
#removing outliers
clean_data = arr[(arr >= lower_bound) & (arr <= upper_bound)]
plt.figure(figsize=(6,4))
sns.boxplot(y=clean_data)
plt.title("Box Plot with Outlier (IQR Method)")
plt.ylabel("Marks")
plt.show()
```
