```python
import matplotlib.pyplot as plt
programming=['java','Python','PHP','JavaScript','C#','C++']
Popularity=[22.2,17.6,8.8,8,7.7,6.7]
plt.title(" popularity of programming language Worldwide ,Oct 2017 compared to year ")
plt.xlabel("programming language")
plt.ylabel("Popularuty")
fig, ax = plt.subplots()
rects1=plt.bar(programming,Popularity,color=['red', 'black', 'green', 'blue', 'yellow', 'cyan'])
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
#plt.grid(which='minor',ls='-',lw='0.5',color='red')
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%f' % float(height),
        ha='center', va='bottom')
autolabel(rects1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
n_groups = 5
fig,ax=plt.subplots()
index=np.arange(n_groups)
bar_width=0.35
opacity=0.8
Means_men = (22, 30, 35, 35, 26)
Means_women = (25, 32, 30, 35, 29)
plt.bar(index, Means_men, bar_width,alpha=opacity,color='g',label='Men')
plt.bar(index+bar_width,Means_women,bar_width,alpha=opacity,color='red',label='Women')
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.legend()
plt.tight_layout()
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
```
