# Data Representation

```python
import numpy as np
import matplotlib.pyplot as pit
import seaborn as sns
#create figure and axis
fig, axs=pit.subplots(3,2,figsize=(14,12))
fig.suptitle("Histograms with KDE of Various Distributions :",fontsize=16)
#Normal Distribution 
data_normal=np.random.normal(loc=0,scale=1,size=1000)
sns.histplot(data_normal,kde=True,ax=axs[0,0],color='blue',stat='density',bins=30)
axs[0,0].set_title('Normal Distribution')
axs[0,0].grid(True)
# Skewd Distribution
data_right_skewd=np.random.exponential(scale=1,size=1000)
data_left_skewd=np.random.exponential(scale=1,size=1000)
# right skewd histograms 
sns.histplot(data_right_skewd,kde=True,ax=axs[0,1],color='red',stat='density',bins=30,label='data_right_skewd')
# left skewd histograms
sns.histplot(data_left_skewd,kde=True,ax=axs[0,1],color='green',stat='density',bins=30,label='data_left_skewd')
axs[0,1].set_title('skewd distribution')
axs[0,1].legend()
axs[0,1].grid(True)
#uniform distribution
uniform_distribution=np.random.uniform(low=0,high=1,size=1000)
sns.histplot(uniform_distribution,kde=True,ax=axs[1,0],color='pink',stat='density',bins=30)
axs[1,0].set_title('Uniform Distribution')
axs[1,0].grid(True)
# biomodel distribution
bi_model=np.concatenate((
    np.random.normal(loc=3,scale=1,size=500),
    np.random.normal(loc=0,scale=1,size=500)))
sns.histplot(bi_model,kde=True,ax=axs[1,1],color='orange',stat='density',bins=30)
axs[1,1].set_title("Biomodel Distribution")
axs[1,1].grid(True)
#multimodel distribution
multi_model=np.concatenate((np.random.normal(loc=0,scale=1,size=500),np.random.normal(loc=3,scale=1,size=500),np.random.normal(loc=6,scale=1,size=500)))
sns.histplot(multi_model,kde=True,ax=axs[2,0],color='green',stat='density',bins=30)
axs[2,0].set_title("Multimode Distribution")
axs[2,0].grid(True)
```
