Range 

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('tips')
max =data.total_bill.max()
min=data.total_bill.min()
print("range :",(max-min))
```

IQR

```python
import numpy as np
Q1=np.percentile(data.total_bill,25)
Q3=np.percentile(data.total_bill,75)
IQR=Q3-Q1
print(IQR)
MEDIAN=np.median(data.total_bill)
MEDIAN
```
